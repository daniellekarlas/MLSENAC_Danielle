##Scatterplot

```python
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,5,6,8,4,8]

"""Usando o marker:
"o" = bolinha
"|" = |
"v" = triangulo
"x" = x
"""

plt.scatter(x,y,label="Pontos", color="r", marker="x", s=80)
plt.legend()
plt.show()
```

##Stack Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]

plt.stackplot(dias, dormir, comer, trabalhar, passear, labels=["dias", "dormir", "comer", "trabalhar", "passear"], colors=["m", "c", "r", "k"])
plt.legend(loc="upper left")
plt.show()
```

##Pie chart ou gráfico de pizza

```python
porcentagens=[70,20,20,130]
cidades = ["São Paulo", "Recife", "Fortaleza", "Manaus"]
colunas = ["c", "m", "r", "k"]
plt.pie(porcentagens, labels= cidades, colors=colunas, startangle=90, shadow=True, explode=(0, 0.15, 0, 0))
plt.savefig("pie_chart.png", dpi=600, size=(100,100))
plt.show()
```

##Gráficos a partir do Numpy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.scatter(np.arange(50), np.random.rand(50), label="Ponto", color="k")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gráfico de pontos criado com Numpy")
```

```python
#plot e scatter

fig=plt.figure()
ax1=fig.add_subplot(1,2,1) #plot
ax1.plot(np.random.randn(50), color="red")

ax2=fig.add_subplot(1,2,2) #scatterplot
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
#controle dos eixos

x=np.linspace(0,1.,100)
fig,axes=plt.subplots(1,2,figsize=(12,4))

axes[0].plot(x,x**2, x, x**3)
axes[0].set_title("Eixos com range padrão")

x=np.linspace(-1, 1., 100)
axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0,1]) #especifica o tamanho do eixo y
axes[1].set_xlim([-1,1]) #especifica o tamanho do eixo x
axes[1].set_title("Eixos customizados")
```

```python
#Diferentes estilos de plots
xx=np.linspace(-0.75,1.,100)
n=np.array([0,1,2,3,4,5])

fig,axes=plt.subplots(1,4, figsize=(12,3))

axes[0].scatter(xx, xx+0.25*np.random.randn(len(xx)))
axes[0].set_title("Scatter")

axes[1].step(n,n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n,n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill between")
```

##Histogramas

```python
n=np.random.randn(100)
fig,axes=plt.subplots(1,2,figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Histograma Padrão")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n),max(n)));



```

```python
#color map (Mapa de calor)
alpha=0.7
phi_ext=2*np.pi*0.5

def ColorMap(phi_m, phi_p):
  return (+ alpha -2 * np.cos(phi_p)*np.cos(phi_m)-alpha*np.cos(phi_ext-2*phi_p))

phi_m=np.linspace(0,2*np.pi,100)
phi_p=np.linspace(0,2*np.pi,100)
X,Y=np.meshgrid(phi_p, phi_m)
Z=ColorMap(X,Y).T

fig, ax=plt.subplots()

p=ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, vmin=abs(Z).min(), vmax=abs(Z).max()) #cmap=cm.RdBu
cb=fig.colorbar(p,ax=ax)
```

##Gráficos 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
#Para conhecer mais sobre o Mplot3d: https://matplotlib.org/2.0.2/mpl_toolkits/mplot3d/tutorial.html
```

```python
fig=plt.figure(figsize=(14,6))

#Um objeto Axes3D é criado como qualquer outro eixo usando a palavra-chave: projection="3d"

ax=fig.add_subplot(1,2,1, projection="3d")
p=ax.plot_surface(X,Y,Z, rstride=4, cstride=4, linewidth=0)

ax=fig.add_subplot(1,2,2, projection="3d")
p=ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb=fig.colorbar(p, shrink=0.5)
```

```python
#Wire frame
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(1,1,1, projection="3d") 
p=ax.plot_wireframe(X,Y,Z, rstride=4, cstride=4)
```

```python
from mpl_toolkits.mplot3d import axes3d
#Countour plot com projeção
fig=plt.figure(figsize=(8,6))

ax=fig.add_subplot(1,1,1, projection="3d")

ax.plot_surface(X,Y,Z, rstride=4, cstride=4, alpha=0.25)
cset=ax.countour(X,Y,Z, zdir="z", offset=-np.pi, cmap=plt.cm.coolwarm)
cset=ax.countour(X,Y,Z, zdir="x", offset=-np.pi, cmap=plt.cm.coolwarm)
cset=ax.countour(X,Y,Z, zdir="y", offset=3*np.pi, cmap=plt.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0,3*np.pi);
ax.set_zlim3d(-np.pi,2*np.pi);
```
