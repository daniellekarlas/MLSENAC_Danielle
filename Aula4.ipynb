Desafio 1: Faa um programa que pea 10 nmeros inteiros, calcule e mostre as quantidades de nmeros pares e a quantidades de nmeros mpares

Desafio 3: Faa um programa que cria uma lista automaticamente usando a funo range () e imprima na tela apenas os nmeros pares. 
Sempre que encontrar um nmero mpar multiplicar por ele mesmo e mostrar na tela.

```python
lista=list(range(0,20))
for numero in lista:
  if numero %2 == 0:
    print(numero)
  else:
    print(numero*numero)
```

Desafio 4: Faa um programa para percorrer uma palavra e imprimir na tela apenas as letras que so vogais e mostrar em qual ndice da palavra cada vogal est associada.
 

```python
palavra=list((input("Digite uma palavra:")))
vogais=["a", "e", "i", "o", "u"]
for indice, letra in enumerate(palavra):
  if letra in vogais:
    print(indice, letra)
```
